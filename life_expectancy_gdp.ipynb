```python
import numpy as np
import pandas as pd
import os
import functools
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
sns.__version__
```

```python
data = pd.read_csv('all_data.csv')

print(data['Country'].unique())

print(data.head())

```

```python
#for one country in the data
chile_data = data[data['Country']=='Chile']
year = chile_data['Year']
lifeex = chile_data['Life expectancy at birth (years)']
GDP = chile_data['GDP']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

ax1.plot(year, lifeex)
ax2.plot(year, GDP)

plt.tight_layout()

plt.show()
```

```python
data1 = data[['Country', 'Year', 'Life expectancy at birth (years)', 'GDP']]
plt.figure(figsize=(5, 30))
plt.subplots_adjust(hspace=0.8)
country = data['Country'].unique()
gdmin = data['GDP'].min()
gdmax = data['GDP'].max()
lemin = data['Life expectancy at birth (years)'].min()
lemax = data['Life expectancy at birth (years)'].max()


# loop through the length of tickers and keep track of index
for n, country in enumerate(country):
    # add a new subplot iteratively
    ax = plt.subplot(6, 2, (2*n)+1)
    
    ax1 = plt.subplot(6, 2, (2*n)+2)
    # filter df and plot ticker on the new subplot axis
    codata = data1[data1["Country"] == country]
    
    ax.plot(codata["Year"], codata["GDP"]) 
    ax.set_title(country.upper())
    ax1.plot(codata["Year"], codata["Life expectancy at birth (years)"])
    ax.set_ylim([gdmin,gdmax])
    ax1.set_ylim([75,lemax])    

plt.show()
```
